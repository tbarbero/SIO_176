```python
T = f.variables['Temperature'].values[:]
P = f.variables['Pressure'].values[:]
t = f.variables['JULD'].values

# These are the variables you will want for these floats 
# NOTE!! You may have to transform t into the right dates using the reference date. Check to make sure your dates 
# look good on the plot
```

```python
tt = pd.DatetimeIndex(t)
# good way to process the time data into readable dates
```

### Section Plot

To avoid interpolation in time and space, which isn't necessary here and would be useless anyway because there is no simple way to interpolate using the one float you guys have - you would have to use other floats as references for the missing data -- instead I did a contour of the property I wanted to look at (temperature) and then relabeled the axes without changing anything 

Relabeling the axes depends on the values that are already being presented. The temperature y-axis is going to be over 400 because that's the number of data points you have. We will be using the actual pressure data but skipping a certain number of points so it can be displayed on the y-axis without overlapping so much you can't read the data

```python
## create a new fake Pres set because it's quick and easy and will convey the same data!

P_lab = []; # open an empty array so you can append to it 

for i in range(0, len(P[1,::-1]), 9): #do a for loop that will skip every x amount of values in the pressure data
    # the line above creates a set of indices that skips every 9th line because the # of total points I have 
    # is divisible by 9 
    pp = P[1,i] 
    
    P_lab = np.append(P_lab,pp) #append each one so you have a pressure set that contains every 9th point
```

In order to figure out how many points you can have on your x and y axis, use T.shape(). This will tell you the number of rows and columns. Should be like (45,558) where there are 45 profiles (45 positions on the time axis) and 558 temperature measurements down the column (558 pressure levels on the y-axis)

```python
fig = plt.figure(1,(15,12))

#create the figure

im = plt.contourf(np.flipud(np.transpose(T)),cmap = 'Blues') #contourf the transposed and flipped data

con = plt.contour(np.flipud(np.transpose(T)),colors='k',levels=20) # add the contours with as many levels as you want

plt.clabel(con, fmt='%4.2f',fontsize=14) # label the contours

plt.colorbar(im) #colorbar is important = add units to it in your own time
positions = range(0,45,1) # this is the range of positions I want for the x-axis. For time, I can use every point on
# the x-axis and just tilt the time labels so they all fit together

labels = (tt.date) # this sets the labels you're going to use for the xaxis equal to just year,month,day
plt.xticks(positions, labels) #this sets the xaxis labels to be the labels defined above at the positions defined above

pos = range(0,558,9) #repeat the same for y-axis -- REMEMBER YOUR DATA MUST SKIP THE SAME # OF INDICES IN LABELS AND POSITIONS
# YOU DONT WANT THEM ALL OR IT WILL BE IMPOSSIBLE TO READ
lab = P_lab
plt.yticks(pos, lab)

degrees = 70 #tilt the date labels 70 degrees so you can see them all
plt.xticks(rotation=degrees);

plt.title('Temperature Section for Argo Float') #label everything
plt.ylabel('Pressure [dbars]')
plt.xlabel('Time')

```
